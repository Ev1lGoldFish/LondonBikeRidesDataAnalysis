```python
#changing column names to understand the data easier
newColsDict = {
  'timestamp':'time',
  'cnt': 'count',
  't1': 'temp_actual_C',
  't2': 'temp_feelsLike_C',
  'hum': 'humidity_percentage',
  'wind_speed': 'windspeed_kmph',
  'weather_code': 'weather',
  'is_holiday': 'is_holiday',
  'is_weekend': 'is_weekend',
  'season': 'season'
}

#renaming the columns in the file to the new names
bikesFile.rename(newColsDict, axis=1, inplace=True)
```

```python
#converting the humidity to percent values
bikesFile.humidity_percentage = bikesFile.humidity_percentage / 100
```

```python
#changing the values of seasons from numbers to actual names
seasonsDict = {
  '0': 'spring',
  '1': 'summer',
  '2': 'autumn',
  '3': 'winter'
}

#changing the datatype of the seasons column from number to string
bikesFile.season = bikesFile.season.astype('str')

#mapping the new values to the season column
bikesFile.season = bikesFile.season.map(seasonsDict)

```

```python
bikesFile.head()
```

```python
#changing the values of weather from numbers to actual names
weatherDict = {
  '1': 'Clear',
  '2':'Scattered Clouds',
  '3': 'Broken Clounds',
  '4': 'Cloudy',
  '7': 'Rain',
  '10': 'Rain with thunderstorm',
  '26': 'Snowfall'
}

#changing the datatype of the weather column from number to string
bikesFile.weather = bikesFile.weather.astype('str')

#mapping the new values to the weather column
bikesFile.weather = bikesFile.weather.map(weatherDict)


```
